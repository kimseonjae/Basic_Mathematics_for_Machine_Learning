# 지도 학습

- 사람이 답을 알려주고 분석을 시키는 것
- MINIST :숫자를 쓰는 방법이 다들 다르기 때문에 여러 사람들의 손글씨를 받은 데이터 셋

- ### 회귀모델(Regression)
        
    - 연속적인 값을 예측
    - 데이터를 통해 특정구간의 값을 구하는 것
    - 직선의 방정식을 정의해 y 절편과 기울기 값을 통해 예측
    - y' = w1x1 + b
        - 예측 값을 y'이라고 표현, y 값은 정답 값
        - w1과 b값은 난수로 처리, b 값과 y' 값을 입력해가면서 정답에 가까워지도록 학습
        - y' = w1x1 + w2x2 + w3x3 + b 복잡해 질수록 w, x 값 추가
    
- ### 분류 모델
    
    - binary classification
        - Pass, Fail을 판단
        - 출력 노드 2개
        
    - Multi-label- classification
        - 어느 범주에 속하는지 분류
        - 클래스 개수가 출력노드의 개수
        
- ### 신경망
    
    - 입력이 들어오면 뉴런에서의 출력을 통해 다음 뉴런에 신호를 전달
        
        - 수상돌기 : 입력을 받아들이는 곳  
        - 축삭돌기 : 다음 뉴런에게 전달해주는 곳
    
    - 신호의 세기에 따라(임계값) 신호를 전달할지 판단
        
        - 입력 신호 없음 x = 0, 입력 신호 있음 x = 1
        - 출력 신호 없음 y = 0, 출력 신호 있음 y = 1
    
    - 활성화 함수(반응 조건 그래프)
        
        - 단위 계단 함수, 미분 불가능
        - 뉴련 반응 여부에 대한 시각화
        - 미분이 가능한 시그모이드 함수로 바꿔서 사용하기도 함
    
    - 입력층, 은닉층, 출력층으로 구성
        
        - 입력층: 이미지의 픽셀의 수와 동일
        - 은닉층: 스스로 학습을 통해 자신들만의 패턴을 만들어가며 가중치를 만들어 판별 
        - 출력층: (클래스의 수)판별해 분류해야 하는 수

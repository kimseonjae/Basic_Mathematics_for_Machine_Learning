```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits # 패키지 임포트

digits = load_digits()
sampleNum = 3 # 출력할 샘플 번호를 지정

img = digits.images[sampleNum]

plt.figure(figsize=(8, 2))
plt.imshow(img, interpolation='nearest', cmap=plt.cm.bone_r) 
#interpolation = 경계선 주변을 무엇으로 채울것인지
plt.grid(False)
plt.xticks([])
plt.yticks([])
plt.title('image {}' .format(sampleNum + 1))
plt.show()
```

```python
digits = load_digits()
samples = [0, 10, 20, 30, 1, 11, 21, 31]
d = []
for i in range(8):
    d.append(digits.images[samples[i]])
plt.figure(figsize=(8, 2))

D = np.array(d)
print(D.shape)

for i in range(8):
    plt.subplot(1, 8, i+1)
    plt.imshow(d[i], interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.title('image {}' .format(i+1))

plt.suptitle("Digit 0 & 1 image")
plt.tight_layout()
plt.show()
```

### fileCategoryCount
- cross, left, right, top 개수 새기


```python
import os

path = "C:\\Users\\rlatj\\Desktop\\AI 융복합\\2단계\\train"
file_list = os.listdir(path)

print("File_ist: {}".format(file_list))
```

```python
crossCount, leftCount, rightCount, topCount = 0, 0, 0, 0


for fileName in file_list:
    dirname, basename = os.path.split(fileName)
    # 파일의 경로는 없애주고 파일 명만 남겨줌
    category = basename.split('_') #category, _ 설명
    # category에는 _ 앞부분을 넣어주고 _ 뒷부분 부터는 저장을 하지 않겠다
    
    if category[0] == 'cross':
        crossCount += 1
    elif category[0] == 'left':
        leftCount += 1
    elif category[0] == 'right':
        rightCount += 1
    else:
        topCount += 1
print('crossCount = {}'.format(crossCount))
print('leftCount = {}'.format(leftCount))
print('rightCount = {}'.format(rightCount))
print('topCount = {}'.format(topCount))
```
